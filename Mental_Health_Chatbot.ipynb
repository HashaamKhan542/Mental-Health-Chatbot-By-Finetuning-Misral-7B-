```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_id = "mistralai/Mistral-7B-Instruct-v0.2"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype="auto"
)

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
prompt = (
    "<s>[INST] You are a supportive mental health assistant. "
    "Respond with empathy and encouragement.\n"
    "User: I've been feeling anxious and overwhelmed with work lately. What should I do? [/INST]"
)
output = pipe(
    prompt,
    max_new_tokens=200,
    do_sample=True,
    top_p=0.9,
    temperature=0.7)
print(output[0]['generated_text'])
```

```python
from datasets import load_dataset
import json

# Step 1: Read and parse each line as a separate dialogue
data = []
with open("augesc.json", "r") as f:
    for line in f:
        try:
            parsed = json.loads(line.strip())
            data.append({"dialog": parsed})
        except json.JSONDecodeError as e:
            print("Skipping a line due to error:", e)
```

```python
from datasets import Dataset

dataset = Dataset.from_list(data)


```

```python
dataset.save_to_disk("augesc_dataset_raw")
```

```python
dataset = dataset.shuffle(seed=42)


train_data = dataset.select(range(0, 7000))
val_data = dataset.select(range(7000, 8000))
test_data = dataset.select(range(8000, 11000))


```

```python
def format_for_mistral(example):
    dialogue = example["dialog"]  # list of ["usr", "text"] and ["sys", "text"]
    formatted = "<s>[INST] You are a supportive mental health assistant.\n"

    for role, text in dialogue:
        if role == "usr":
            formatted += f"User: {text}\n"
        elif role == "sys":
            formatted += f"Assistant: {text}\n"

    formatted += "[/INST]"
    return {"text": formatted}
```

```python

train_data = train_data.map(format_for_mistral, remove_columns=["dialog"])
val_data = val_data.map(format_for_mistral, remove_columns=["dialog"])
test_data = test_data.map(format_for_mistral, remove_columns=["dialog"])
```

# Benchmarking the Base model (First 100 Samples)

```python
val_data = load_from_disk("/content/drive/MyDrive/Mental_Health_Chatbot/augesc_dataset_splits/val")
val_texts = val_data["text"][:100]
```

```python
prompts = []
references = []

for text in val_texts:
    if "User:" in text and "Assistant:" in text:
        turns = text.split("\n")
        convo = [t.strip() for t in turns if t.strip().startswith(("User:", "Assistant:"))]

        try:
            first_user_idx = next(i for i, line in enumerate(convo) if line.startswith("User:"))
            prompt = "\n".join(["<s>[INST] You are a supportive mental health assistant."] + convo[:first_user_idx+1])
            ref = convo[first_user_idx+1] if first_user_idx+1 < len(convo) else ""

            if ref.startswith("Assistant:"):
                prompts.append(prompt + "\n[/INST]")
                references.append(ref.replace("Assistant:", "").strip())
        except StopIteration:
            continue

```

```python
tokenizer.pad_token = tokenizer.eos_token

generated = []

for prompt in prompts:
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True, max_length=1024).to("cuda")
    output = model.generate(**inputs, max_new_tokens=128, do_sample=True, top_k=50, top_p=0.95)

    decoded = tokenizer.decode(output[0], skip_special_tokens=True)

    # Extract just the assistant's reply
    if "Assistant:" in decoded:
        reply = decoded.split("Assistant:")[-1].strip().split("\n")[0]
    else:
        reply = decoded.strip()

    generated.append(reply)
```

```python
generated_dataset = Dataset.from_dict({"text": generated})
generated_dataset.save_to_disk("/content/drive/MyDrive/Mental_Health_Chatbot/augesc_dataset_splits/val/Output_For_First_100/")
```

```python
from bert_score import score

# generated = [...]  # Your base model generated responses
# references = [...]  # Ground truth responses

assert len(generated) == len(references), "Mismatch in # of predictions and references"

# Compute BERTScore
P, R, F1 = score(generated, references, lang="en")

# Print average scores
print(f"\n🔍 BERTScore (P):   {P.mean().item():.4f}")
print(f"🔍 BERTScore (R):   {R.mean().item():.4f}")
print(f"🔍 BERTScore (F1):  {F1.mean().item():.4f}  ✅ (Main Metric)")
```

```python
import pandas as pd

df = pd.DataFrame({
    "prompt": prompts[:len(generated)],
    "reference": references[:len(generated)],
    "generated": generated,
    "BERTScore_P": P.tolist(),
    "BERTScore_R": R.tolist(),
    "BERTScore_F1": F1.tolist()
})

df.to_csv("/content/drive/MyDrive/Mental_Health_Chatbot/val/bert_scores_base_model.csv", index=False)
```
